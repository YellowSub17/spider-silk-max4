```python
import numpy as np


Is = []
scans = [ 
    Scan(102809, load_n=1), #YRA in kpi cap1 sample1
    Scan(102815, load_n=1), #YRA in kpi cap1 sample1 
    Scan(102821, load_n=1), #YRA in MQ cap2 sample1
    Scan(102827, load_n=1), #YRA in MQ cap2 sample1
    Scan(102824, load_n=1), #kpi cap2 sample4
    Scan(102830, load_n=1), #kpi cap2 sample4
]

scans = [ 
    Scan(i, load_n=1) for i in range(102809, 102845)
]

plt.figure()

for s in scans:
    i=0
    # --- 1. Define your boundaries (from your red dashed lines) ---
    q_start = det.DET_OVERLAP[0] # Where overlap begins
    q_end = det.DET_OVERLAP[1]    # Where overlap ends
    
    # --- 2. Define a master q-grid for the final stitched curve ---
    # We take SAXS points up to q_start, and WAXS points after q_end.
    # In the middle, we can use either grid (let's use the SAXS density for the overlap).
    q_overlap = s.qs_saxs[(s.qs_saxs >= q_start) & (s.qs_saxs <= q_end)]
    q_combined = np.concatenate([
        s.qs_saxs[s.qs_saxs < q_start],
        q_overlap,
        s.qs_waxs[s.qs_waxs > q_end]
    ])
    q_combined = np.unique(q_combined) # Ensure sorted and unique
    
    # --- 3. Interpolate both curves onto the master q-grid ---
    # This ensures they share the exact same x-coordinates for easy blending.
    I_saxs_interp = np.interp(q_combined, s.qs_saxs, s.Is_saxs[i])
    
    # Use your calculated scale factor on the WAXS curve
    I_waxs_scaled = s.Is_waxs[i] * s.saxswaxs_sf[i]
    I_waxs_interp = np.interp(q_combined, s.qs_waxs, I_waxs_scaled)
    
    # --- 4. Blend them together using a linear ramp ---
    I_combined = np.zeros_like(q_combined)
    
    # Region 1: Pure SAXS
    mask_pure_saxs = q_combined < q_start
    I_combined[mask_pure_saxs] = I_saxs_interp[mask_pure_saxs]
    
    # Region 2: Pure WAXS
    mask_pure_waxs = q_combined > q_end
    I_combined[mask_pure_waxs] = I_waxs_interp[mask_pure_waxs]
    
    # Region 3: Overlap Blend
    mask_overlap = (q_combined >= q_start) & (q_combined <= q_end)
    q_overlap_pts = q_combined[mask_overlap]
    
    # Calculate weights: 1 at q_start (100% SAXS), 0 at q_end (100% WAXS)
    w_saxs = (q_end - q_overlap_pts) / (q_end - q_start)
    w_waxs = 1.0 - w_saxs
    
    
    x_norm = (q_overlap_pts - q_start) / (q_end - q_start)
    k = 5  
    w_waxs = 1 / (1 + np.exp(-k * (x_norm - 0.2)))  # Centered at 20% into the overlap region
    w_saxs = 1.0 - w_waxs
    
    I_combined[mask_overlap] = (w_saxs * I_saxs_interp[mask_overlap]) + (w_waxs * I_waxs_interp[mask_overlap])
    
    # --- 5. Plot your beautifully stitched curve ---

    Is.append(I_combined)
    plt.plot(q_combined, I_combined, label=s.scan_id,  linestyle='-', marker='', markersize=1)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('q')
plt.ylabel('Intensity')
#plt.legend()
plt.show()
```
