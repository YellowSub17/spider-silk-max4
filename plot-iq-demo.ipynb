```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dat_files =[
            '../data/visitors/cosaxs/20241304/Process/scan-27870_eiger.dat',
            '../data/visitors/cosaxs/20241921/reprocess/scan-39495_pilatus.dat',
            '../data/visitors/cosaxs//20241921/process/scan-39524_eiger_sub.dat',
            '../data/visitors/cosaxs//20241921/process/scan-39517_pilatus.dat',
            '../data/visitors/cosaxs//20241921/process/scan-39529_eiger.dat',
            '../data/visitors/cosaxs//20251747/process/scan-39529_eiger.dat'
           ]
```

```python
def open_dat(fname):
    q, I, errors = np.loadtxt(fname).T
    return q, I, errors
    
for dat_file in dat_files:
    q, I, errors = open_dat(dat_file)
    plt.figure()
    plt.title(dat_file)
    #plt.plot(q,I)
    plt.errorbar(q,I, yerr=errors, capsize=1, linestyle='')
```
